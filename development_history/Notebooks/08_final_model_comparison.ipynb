```python
#merging both models arima and sarimax for comparisons
#model evaluations: arima gave accurate values of MAPE for 6/8 chemicals but sarimax gave better results as in prediction values<10% or error for 6/8 chemicals.
import pandas as pd
import numpy as np
```

```python
arima_df = pd.read_csv("/Users/mangalyaphaye/Desktop/suvira_intelligence/outputs/final_arima_results.csv")
sarimax_df = pd.read_csv("/Users/mangalyaphaye/Desktop/suvira_intelligence/outputs/final_sarimax_results.csv")

arima_df.head(), sarimax_df.head()
```

```python
arima_df = arima_df.rename(columns={
    "MAE": "ARIMA_mae",
    "RMSE": "ARIMA_rmse",
    "MAPE": "ARIMA_mape"
})

sarimax_df = sarimax_df.rename(columns={
    "MAE": "SARIMAX_mae",
    "RMSE": "SARIMAX_rmse",
    "MAPE": "SARIMAX_mape"
})

arima_df.head(), sarimax_df.head()
```

```python
comparison_df = pd.merge(arima_df, sarimax_df, on="Chemical")
comparison_df
```

```python
comparison_df["Best Model"] = np.where(comparison_df["SARIMAX_mape"] < comparison_df["ARIMA_mape"],"SARIMAX","ARIMA")
comparison_df
```

```python
#creating a unified pipeline to automate choosing which prediction model is better for which chemical according to it's pattern
comparison_df.to_csv("/Users/mangalyaphaye/Desktop/suvira_intelligence/outputs/model_comparison.csv", index=False)
```
