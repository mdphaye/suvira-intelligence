```python
import pandas as pd
import os
```

```python
folder = "/Users/mangalyaphaye/Desktop/suvira_intelligence/data/processed"

files = [
"urea_monthly.csv",
"kcl_monthly.csv",
"sulphur_monthly.csv",
"sulphuric_acid_monthly.csv",
"acetic_acid_monthly.csv",
"pac_monthly.csv",
"formates_monthly.csv",
"wtc_monthly.csv"
]

dfs = []

for f in files:
    df = pd.read_csv(os.path.join(folder, f))
    dfs.append(df)

master_df = pd.concat(dfs, ignore_index=True)
```

```python
#January
jan_2026 = pd.DataFrame([
["2026-01-01", None, None, None, "Urea"],
["2026-01-01", None, None, None, "KCL"],
["2026-01-01", None, None, None, "Sulphur"],
["2026-01-01", None, None, None, "Sulphuric Acid"],
["2026-01-01", None, None, None, "Acetic Acid"],
["2026-01-01", None, None, None, "PAC"],
["2026-01-01", None, None, None, "Formates"],
["2026-01-01", None, None, None, "Water Treatment Chemicals"]
], columns=master_df.columns)

jan_2026
```

```python
#updating the spot values manually from spot values of the chemicals. (*Ask Tushar sir)
```

```python
updates = [
("2026-01-01", "Urea", 520),
("2026-01-01", "Sulphur", 430),
("2026-01-01", "Acetic Acid", 660),

("2026-02-01", "Urea", 530),
("2026-02-01", "Sulphur", 450),
("2026-02-01", "Acetic Acid", 700)
]

for date, chem, price in updates:
    master_df.loc[
        (master_df["date"] == date) & (master_df["chemical"] == chem),
        "price_usd_per_ton"
    ] = price
```
