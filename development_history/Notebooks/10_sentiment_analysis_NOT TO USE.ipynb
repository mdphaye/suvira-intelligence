```python
#integrating sentiment analysis from news data for prediction of prices based on war situation, etc.
#Simulated pipeline for now
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/mangalyaphaye/Desktop/suvira_intelligence/data/final_2020_2026.csv")
df["date"] = pd.to_datetime(df["date"])
```

```python
np.random.seed(42)

# df["sentiment_score"] = np.random.uniform(-1, 1, size=len(df))

#Event-driven sentiment logic
df["sentiment_score"] = 0

for i in range(len(df)):
    
    date = df.loc[i, "date"]
    
    #COVID period (2020 - early 2021): strongly negative
    if date >= pd.to_datetime("2020-03-01") and date <= pd.to_datetime("2021-03-01"):
        df.loc[i, "sentiment_score"] = np.random.uniform(-0.9, -0.5)
    
    #Recovery phase (2021–2022): positive
    elif date > pd.to_datetime("2021-03-01") and date <= pd.to_datetime("2022-06-01"):
        df.loc[i, "sentiment_score"] = np.random.uniform(0.2, 0.7)
    
    #Volatility phase (2022–2024): mixed
    elif date > pd.to_datetime("2022-06-01") and date <= pd.to_datetime("2024-12-01"):
        df.loc[i, "sentiment_score"] = np.random.uniform(-0.3, 0.3)
    
    #Stable/low demand (2025): slightly negative
    elif date > pd.to_datetime("2024-12-01") and date <= pd.to_datetime("2026-01-31"):
        df.loc[i, "sentiment_score"] = np.random.uniform(-0.2, 0.1)
    
    #2026 Iran conflict/supply shock: negative/high volatility
    else:
        df.loc[i, "sentiment_score"] = np.random.uniform(-0.8, -0.2)
```

```python
#randomness factor
df["sentiment_score"] += np.random.normal(0, 0.05, size=len(df))
```
