```python
# improved model (SARIMAX)

import pandas as pd
import numpy as np

df = pd.read_csv("/Users/mangalyaphaye/Desktop/suvira_intelligence/data/final_2020_2026_with_sentiment.csv")

df["date"] = pd.to_datetime(df["date"])

df.head()
```

```python
# Adding macro drivers (vectorized + cleaner)

df["crude_oil_price"] = np.where(
    df["date"] < "2022-02-01", 70,
    np.where(df["date"] < "2024-01-01", 90, 110)
)

df["natural_gas_price"] = np.where(
    df["date"] < "2022-02-01", 2,
    np.where(df["date"] < "2024-01-01", 3.5, 5)
)

df["war_index"] = np.where(
    df["date"] >= "2022-02-01", 1, 0
)

df.head()
```

```python
#Filter Urea data (brute force, taking one chemical first)
#Model training

from statsmodels.tsa.statespace.sarimax import SARIMAX

urea_df = df[df["chemical"] == "Urea"].copy()
urea_df = urea_df.sort_values("date")

# TRAIN /TEST SPLIT (same as ARIMA)
train = urea_df[urea_df["date"] <= "2025-08-01"].copy()
test  = urea_df[(urea_df["date"] >= "2025-09-01") & 
                (urea_df["date"] <= "2025-12-01")].copy()

# Set index
train.set_index("date", inplace=True)
test.set_index("date", inplace=True)

# Target variable
y_train = train["price_usd_per_ton"].asfreq("MS")
y_test  = test["price_usd_per_ton"].asfreq("MS")

# Exogenous variables
X_train = train[[
    "crude_oil_price",
    "natural_gas_price",
    "war_index",
    "sentiment_score"
]].asfreq("MS")

X_test = test[[
    "crude_oil_price",
    "natural_gas_price",
    "war_index",
    "sentiment_score"
]].asfreq("MS")

# Drop nulls (important alignment)
y_train = y_train.dropna()
X_train = X_train.loc[y_train.index]

y_test = y_test.dropna()
X_test = X_test.loc[y_test.index]

# SARIMAX model
model = SARIMAX(
    y_train,
    exog=X_train,
    order=(1,1,1),
    enforce_stationarity=False,
    enforce_invertibility=False
)

model_fit = model.fit()

print(model_fit.summary())
```

```python
forecast = model_fit.forecast(steps=len(y_test), exog=X_test)

print("Forecast:")
print(forecast)

print("\nActual:")
print(y_test)
```

```python
#evaluation
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(y_test, forecast)
rmse = np.sqrt(mean_squared_error(y_test, forecast))
mape = np.mean(np.abs((y_test - forecast) / y_test)) * 100

print("\nMAE:", mae)
print("RMSE:", rmse)
print("MAPE:", mape)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

chemicals = df["chemical"].unique()

results = []

for chem in chemicals:
    
    temp_df = df[df["chemical"] == chem].copy()
    temp_df = temp_df.sort_values("date")
    
    # TRAIN /TEST SPLIT
    train = temp_df[temp_df["date"] <= "2025-08-01"].copy()
    test  = temp_df[(temp_df["date"] >= "2025-09-01") & 
                    (temp_df["date"] <= "2025-12-01")].copy()
    
    train.set_index("date", inplace=True)
    test.set_index("date", inplace=True)
    
    y_train = train["price_usd_per_ton"].asfreq("MS")
    y_test  = test["price_usd_per_ton"].asfreq("MS")
    
    X_train = train[[
        "crude_oil_price",
        "natural_gas_price",
        "war_index",
        "sentiment_score"
        ]].asfreq("MS")

    X_test = test[[
        "crude_oil_price",
        "natural_gas_price",
        "war_index",
        "sentiment_score"
        ]].asfreq("MS")
    
    # Align
    y_train = y_train.dropna()
    X_train = X_train.loc[y_train.index]
    
    y_test = y_test.dropna()
    X_test = X_test.loc[y_test.index]
    
    # Skip if insufficient data
    if len(y_train) < 20 or len(y_test) == 0:
        continue
    
    try:
        model = SARIMAX(
            y_train,
            exog=X_train,
            order=(1,1,1),
            enforce_stationarity=False,
            enforce_invertibility=False
        )
        
        model_fit = model.fit()
        
        # Forecast TEST (NOT future)
        forecast = model_fit.forecast(steps=len(y_test), exog=X_test)
        
        # Metrics
        mae = mean_absolute_error(y_test, forecast)
        rmse = np.sqrt(mean_squared_error(y_test, forecast))
        mape = np.mean(np.abs((y_test - forecast) / y_test)) * 100
        
        results.append({
            "Chemical": chem,
            "MAE": mae,
            "RMSE": rmse,
            "MAPE": mape
        })
        
    except:
        continue

# Results table
results_df = pd.DataFrame(results)
results_df = results_df.sort_values("MAPE")

results_df
```

```python
results_df.to_csv("sarimax_results.csv", index=False)
```
