```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = "/Users/mangalyaphaye/Desktop/suvira_intelligence/data/raw/urea_raw_combined(2020-2024_.csv"

df = pd.read_csv(file_path, encoding="latin-1", index_col=False)

df.head()
```

```python
df.columns
```

```python
df_clean = df[[
    "refYear",
    "refMonth",
    "reporterISO",
    "reporterDesc",
    "cmdCode",
    "cmdDesc",
    "netWgt",
    "fobvalue"
]].copy()

df_clean = df_clean[df_clean["netWgt"] > 1000]
df_clean.head()
```

```python
monthly_trade = df_clean.groupby(["refYear", "refMonth"]).agg({"netWgt": "sum","fobvalue": "sum"}).reset_index()
monthly_trade.head()
```

```python
monthly_trade["price_usd_per_ton"] = (monthly_trade["fobvalue"] / monthly_trade["netWgt"]) * 1000
monthly_trade.head()
```

```python
monthly_trade["date"] = pd.to_datetime(
    monthly_trade["refYear"].astype(str) + "-" +
    monthly_trade["refMonth"].astype(str)
)

monthly_trade.head()
```

```python
monthly_trade = monthly_trade.sort_values("date").reset_index(drop=True)
```

```python
monthly_trade = monthly_trade.set_index("date")
monthly_trade.head()
```

```python
monthly_trade.index
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(monthly_trade.index, monthly_trade["price_usd_per_ton"], marker='o')

plt.title("Global Acetic Acid Price Trend 2025 (USD/ton)")
plt.xlabel("Month")
plt.ylabel("Price (USD/ton)")

plt.xticks(monthly_trade.index, monthly_trade.index.strftime('%b'),)

plt.grid(True)
plt.show()
```

```python
monthly_trade["chemical"] = "Sulphuric Acid"
monthly_trade.head()
```

```python
monthly_trade = monthly_trade.drop(columns=["refYear", "refMonth"])
monthly_trade.head()
```

```python
monthly_trade = monthly_trade.reset_index()
monthly_trade.head()
```

```python
monthly_trade.to_csv("/Users/mangalyaphaye/Desktop/suvira_intelligence/data/processed/sulphuric_acid_monthly.csv", index=False)
```
