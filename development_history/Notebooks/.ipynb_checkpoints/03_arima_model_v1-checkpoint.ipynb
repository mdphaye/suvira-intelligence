```python
#baseline model
import pandas as pd
import numpy as np

df = pd.read_csv("/Users/mangalyaphaye/Desktop/suvira_intelligence/data/master_dataset_2026.csv")

df.head()
```

```python
#Keep only required columns and sorting values(for time series)
df_ml = df[["date", "chemical", "price_usd_per_ton"]].copy()

df_ml["date"] = pd.to_datetime(df_ml["date"])


df_ml = df_ml.sort_values(["chemical", "date"])

df_ml.head()
```

```python
#Taking one chemical first, example: urea
```

```python
#filter data for one chemical(Urea)
urea_df = df_ml[df_ml["chemical"] == "Urea"].copy()
urea_df.head()
```

```python
# Set date as index:for ARIMA
# urea_df.set_index("date", inplace=True)
urea_ts = urea_df["price_usd_per_ton"]
urea_ts = urea_ts.asfreq('MS') 
urea_ts.head()
```

```python
from statsmodels.tsa.arima.model import ARIMA

#training ARIMA model
model = ARIMA(urea_ts, order=(1,1,1))  #basic configuration
model_fit = model.fit()

print(model_fit.summary())
```

```python
#Predict next 1 month (urea)
forecast = model_fit.forecast(steps=1)
print("Predicted price for next month:")
print(f"Predicted Urea Price for March 2026: {forecast.values[0]:.2f} USD/ton")
```

```python
#Made an automated code for the above process just like what I did in urea
#applied for all the chemicals in the dataset for prediction of march 2026 values.
```

```python
chemicals = df_ml["chemical"].unique()

predictions = {}

for chem in chemicals:
    
    temp_df = df_ml[df_ml["chemical"] == chem].copy()
    temp_df.set_index("date", inplace=True)
    
    ts = temp_df["price_usd_per_ton"]
    ts = ts.asfreq('MS')
    
    ts = ts.dropna()
    
    model = ARIMA(ts, order=(1,1,1))
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=1) 
    predictions[chem] = forecast.values[0]

for chem, value in predictions.items():
    print(f"{chem}: {value:.2f} USD/ton")
```

```python
#tabular output
pred_df = pd.DataFrame(list(predictions.items()), columns=["Chemical", "Predicted Price (USD/ton)"])
pred_df["Predicted Price (USD/ton)"] = pred_df["Predicted Price (USD/ton)"].round(2)
pred_df
```
